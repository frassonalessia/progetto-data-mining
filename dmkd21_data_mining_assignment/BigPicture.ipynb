```python
import graphviz as GV
def to_dot(self):
    return str(self)

GV.Digraph.to_dot =  to_dot
```

```python
eg  = GV.Digraph()
eg.graph_attr['rankdir'] = 'TB'  
eg.node('flow_start','', shape='circle')
eg.node('data_fitbit','Fitbit\n Dataset', shape='folder', )
eg.node('task_etl','t1 - Extract Transform Loading\n Relevant Data', shape='box', peripheries='2')

eg.node('data_selected','Selected RAW Data\n (DataFrames and/or .csv files)', shape='component')
eg.node('task_profiling_aggregate','t2 -Profiling & Aggregate Data', shape='box', peripheries='2')

eg.node('data_events','Event Data \n (DataFrames and/or .csv files)', shape='component')
eg.node('data_timeseries','Timeseries Tables \n (DataFrames and/or .csv files)', shape='component')

eg.node('task_timeseries_to_events','t3 - Transform Timeseries\n into Events', shape='box', peripheries='2')

eg.node('data_discretized_timeseries','Discretized Timeseries Tables \n (DataFrames and/or .csv files)', shape='component')

eg.node('task_feature_reduction', 't4 - Feature Reduction for Events\n (i.e., Reducing the Cardinality of Labels)', shape='box', peripheries='2')

eg.node('data_output_events', 'Output Events\n (DataFrames and/or .csv files)', shape='component')

eg.node('task_transactions', 't5 - Make Transactions', shape='box', peripheries='2')

eg.node('data_transactions', ' Transactions\n (DataFrames and/or .csv files and/or .xes files)', shape='component')

eg.node('task_tarm', 't6 - Temporal Association Rules\n Mining', shape='box', peripheries='2')
eg.node('task_pd', 't7 - Process Discovery', shape='box', peripheries='2')

eg.node('data_tar', 'Temporal Association Rules', shape='component')
eg.node('data_process', 'Process Diagram', shape='component')

eg.node('flow_end', '',  shape='doublecircle')

eg.node('note_profiling_aggregate','parametrization required\n(e.g., granularity of aggregation)', shape='box', style='dashed')
eg.node('note_timeseries_to_events','parametrization required\n(e.g., discretization parameters \n such as time window frame)',shape='box', style='dashed')
eg.node('note_feature_reduction','parametrization required\n(e.g., parameters such as number \n of clusters per label)', shape='box', style='dashed')
eg.node('note_transactions','parametrization required\n(e.g., parameters such as time frame\n size for aggregate events in\n a transaction like days/weeks)', shape='box', style='dashed')
eg.node('note_tarm','parametrization required\n(e.g., parameters related to\n temporal association rules mining\n such as support/confidence\n thresholds and so on)', shape='box', style='dashed')

###EDGES

eg.edge('flow_start', 'task_etl')
eg.edge('data_fitbit', 'task_etl')
eg.edge('task_etl', 'data_selected')

eg.edge('data_selected', 'task_profiling_aggregate')

eg.edge('task_profiling_aggregate', 'data_events')
eg.edge('task_profiling_aggregate', 'note_profiling_aggregate', dir = 'none', style='dashed')


eg.edge('task_profiling_aggregate', 'data_timeseries')
eg.edge('data_timeseries', 'task_timeseries_to_events')

eg.edge('task_timeseries_to_events', 'data_discretized_timeseries')
eg.edge('task_timeseries_to_events', 'note_timeseries_to_events', dir = 'none', style='dashed')

eg.edge('data_discretized_timeseries', 'task_feature_reduction')
eg.edge('data_events', 'task_feature_reduction')

eg.edge('task_feature_reduction','data_output_events')
eg.edge('task_feature_reduction','note_feature_reduction', dir = 'none', style='dashed')

eg.edge('data_output_events', 'task_transactions')
eg.edge('task_transactions', 'data_transactions')
eg.edge('task_transactions','note_transactions', dir = 'none', style='dashed')


eg.edge('data_transactions', 'task_tarm')
eg.edge('data_transactions', 'task_pd')

eg.edge('task_tarm', 'data_tar')
eg.edge('task_tarm','note_tarm', dir = 'none', style='dashed')

eg.edge('task_pd', 'data_process')

eg.edge('data_tar', 'flow_end')
eg.edge('data_process', 'flow_end')

%dotobj eg
```
